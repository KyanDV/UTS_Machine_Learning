```python
#Kyan D.V. - 223400001
#MULTIPLE LINEAR REGRESSION
import pandas as pd
df = pd.read_excel('/content/HARGA RUMAH JAKSEL.xlsx')
df
```

```python
df['KOTA'].value_counts()
```

```python
df['GRS'] = df['GRS'].map({'ADA': 1, 'TIDAK ADA': 0})
df
```

```python
df['GRS'].value_counts()
```

```python
df['KOTA'] = df['KOTA'].map({'JAKSEL': 0})
df
```

```python
import statsmodels.api as sm
Y = df['HARGA']
X = df[['LT', 'LB', 'JKT', 'JKM', 'GRS', 'KOTA']]
X = sm.add_constant(X)
```

```python
ks = sm.OLS(Y, X)
ks_res =ks.fit()
ks_res.summary()
```

```python
#POLYNOMIAL REGRESSION
import datetime as dt
import sys
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from math import sqrt
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

poly_reg = PolynomialFeatures(degree=2) # You can adjust the degree
X_poly_train = poly_reg.fit_transform(X_train)
X_poly_test = poly_reg.transform(X_test)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_train, y_train)
y_pred_poly = lin_reg_2.predict(X_poly_test)

rmse_poly = sqrt(mean_squared_error(y_test, y_pred_poly))
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = lin_reg_2.score(X_poly_test, y_test)


y_pred_mlr = ks_res.predict(X_test)
rmse_mlr = sqrt(mean_squared_error(y_test, y_pred_mlr))
mse_mlr = mean_squared_error(y_test, y_pred_mlr)
r2_mlr = ks_res.rsquared

print("Multiple Linear Regression:")
print(f"R-squared: {r2_mlr}")
print(f"RMSE: {rmse_mlr}")
print(f"MSE: {mse_mlr}")

print("\nPolynomial Regression:")
print(f"R-squared: {r2_poly}")
print(f"RMSE: {rmse_poly}")
print(f"MSE: {mse_poly}")
```
